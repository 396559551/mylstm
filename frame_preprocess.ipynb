```python
import os
import pickle
import numpy as np
import random
np.set_printoptions(suppress=True)
```

```python
batch_size=48
seq_length=1
datasets=[0, 1, 2, 3, 4]
forcePreProcess=False
infer=False
data_dirs = ['./data/eth/univ', './data/eth/hotel',
                  './data/ucy/zara/zara01', './data/ucy/zara/zara02',
                  './data/ucy/univ']
used_data_dirs = [data_dirs[x] for x in datasets]
infer = infer

# Number of datasets
numDatasets = len(data_dirs)

# Data directory where the pre-processed pickle file resides
data_dir = './data'

# Store the arguments
batch_size = batch_size
seq_length = seq_length

# Validation arguments
val_fraction = 0.2
```

```python
data_file = os.path.join(data_dir, "trajectories_1.pkl")
```

```python
data_dirs=used_data_dirs
```

```python
all_frame_data = []
valid_frame_data = []
frameList_data = []
numPeds_data = []
dataset_index = 0
```

```python
for directory in data_dirs:
    file_path = os.path.join(directory, 'pixel_pos_interpolate_1.csv')
    data = np.genfromtxt(file_path, delimiter=',')
    frameList = np.unique(data[0, :]).tolist()
    numFrames = len(frameList)
    frameList_data.append(frameList)
    numPeds_data.append([])
    all_frame_data.append([])
    valid_frame_data.append([])
    skip = 10
    for ind, frame in enumerate(frameList):
        if ind % skip != 0:
            continue
        pedsInFrame=data[:, data[0, :] == frame]
        pedsList = pedsInFrame[1, :].tolist()
        numPeds_data[dataset_index].append(pedsList)
        pedsWithPos = []
        for ped in pedsList:
            current_x = pedsInFrame[3, pedsInFrame[1, :] == ped][0]
            current_y = pedsInFrame[2, pedsInFrame[1, :] == ped][0]
            pedsWithPos.append([ped, current_x, current_y])
        if (ind > numFrames * val_fraction) or (infer):
            all_frame_data[dataset_index].append(np.array(pedsWithPos))
        else:
            valid_frame_data[dataset_index].append(np.array(pedsWithPos))
    dataset_index += 1
f = open(data_file, "wb")
pickle.dump((all_frame_data, frameList_data, numPeds_data, valid_frame_data), f, protocol=2)
f.close()
```
